##Importação da biblioteca

```python
# Importação da biblioteca principal para análise de dados

import pandas as pd

# carregar arquivo (acidentes2025_todas_causas_tipos)

df = pd.read_csv(
    'acidentes2025_todas_causas_tipos.csv',   # Leitura do arquivo CSV contendo os dados de acidentes
    sep=';',                                  # sep=',' → separador das colunas (vírgula)
    encoding='latin1',                        # encoding='latin1' → evita erros com caracteres especiais
    low_memory=False                          # low_memory=False → melhora a leitura de colunas com tipos mistos
)

# visualizar dados
df.head()
```

## Entendimento dos dados

```python
# Exibe o nome de todas as colunas do dataset
print("Colunas:")
print(df.columns)

# Mostra informações gerais:
# - quantidade de linhas
# - tipos de dados
# - valores não nulos
print("\nInformações:")
df.info()

# Exibe estatísticas descritivas para colunas numéricas
print("\nResumo estatístico:")
df.describe()
```

## Limpeza inicial

```python
# valores nulos
# Identifica a quantidade de valores nulos por coluna
# Ordena do maior para o menor
df.isnull().sum().sort_values(ascending=False).head(15)
```

```python
# remover duplicados
# Remove linhas duplicadas do dataset
# Isso evita distorções na análise

df = df.drop_duplicates()
```

#Tratamento inicial

## Converter data

```python
# Converter para formato de data
df['data_inversa'] = pd.to_datetime(df['data_inversa'])
```

## Extrair hora

```python
# Extrair apenas a hora (0–23)
df['hora'] = pd.to_datetime(df['horario']).dt.hour
```

## Converter latitude/longitude

```python
# Converter valores de texto para numérico
df['latitude'] = pd.to_numeric(df['latitude'], errors='coerce')
df['longitude'] = pd.to_numeric(df['longitude'], errors='coerce')
```

## variável de gravidade

```python
# Criar variável de gravidade
df['gravidade'] = df['mortos'] + df['feridos_graves']
```

#VISUALIZAÇÃO



## Gráfico de acidentes por hora

```python
import matplotlib.pyplot as plt

df['hora'].value_counts().sort_index().plot(kind='bar')

plt.title('Acidentes por Hora')
plt.xlabel('Hora do dia')
plt.ylabel('Quantidade de acidentes')
plt.show()
```

##Top 10 estados

```python
df['uf'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Estados com Mais Acidentes')
plt.xlabel('Estado')
plt.ylabel('Quantidade')
plt.show()
```

## Principais causas

```python
df['causa_acidente'].value_counts().head(10).plot(kind='barh')

plt.title('Principais Causas de Acidentes')
plt.xlabel('Quantidade')
plt.show()
```

## Gravidade por estado

```python
df.groupby('uf')['gravidade'].sum().sort_values(ascending=False).head(10).plot(kind='bar')

plt.title('Gravidade dos Acidentes por Estado')
plt.ylabel('Gravidade')
plt.show()
```

# Análises Avançadas

## Taxa de gravidade por estado

```python
df_estado = df.groupby('uf').agg({
    'id': 'count',
    'gravidade': 'sum'
})

df_estado['taxa_gravidade'] = df_estado['gravidade'] / df_estado['id']

df_estado.sort_values('taxa_gravidade', ascending=False).head(10)
```

## Acidentes por dia da semana

```python
df['dia_semana'].value_counts()
```

```python
ordem_dias = [
    'segunda-feira', 'terça-feira', 'quarta-feira',
    'quinta-feira', 'sexta-feira', 'sábado', 'domingo'
]

df['dia_semana'] = pd.Categorical(df['dia_semana'], categories=ordem_dias, ordered=True)

df['dia_semana'].value_counts().sort_index().plot(kind='bar')
```

## Gravidade por período do dia

```python
df.groupby('fase_dia')['gravidade'].sum().sort_values(ascending=False)
```

## Clima vs gravidade:
